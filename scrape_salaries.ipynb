```python
from robobrowser import RoboBrowser
import pandas as pd
from io import StringIO
```

```python
ROTOGURU_STUB = "http://rotoguru1.com/cgi-bin/fyday.pl"
```

```python
def scrape_salaries(br, year, week):
    br.open(ROTOGURU_STUB + "?week={}&year={}&game=dk&scsv=1".format(week, year))
    table = br.find('pre')
    df = pd.read_csv(StringIO(table.contents[0]), sep=';')
    return df
```

```python
br = RoboBrowser(history=True)
full_df = pd.DataFrame()
for year in [str(y) for y in range(2016, 2018)]:
    for week in [str(x) for x in range(1, 18)]:
        df = scrape_salaries(br, year, week)
        full_df = full_df.append(df, ignore_index=True)
full_df.to_csv('historical_salaries.csv', index=False)
```
